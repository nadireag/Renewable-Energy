```python
df = pd.read_csv("db/us_comparison.csv", index_col=0)
df.head()
```

```python
X = df["Year"].values.reshape(-1,1)
y = df["Comparison"].values.reshape(-1,1)
# print the first 5 year values
print(X[:5])
```

```python
# Now we implement a 5-fold CV procedure.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score


predicted = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
# Neural Network
# Deep learning algorithms with keras
from keras.layers import Dense
from keras.models import Sequential
from keras import optimizers
```

```python
from keras import backend as K

def R2(y_true, y_pred):
    u = K.sum(K.square( y_true-y_pred ))
    v = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - u/(v + K.epsilon()) )
```

```python
X = df.drop(['Prod Data','Cons Data','Status','Comparison'], axis=1)
X.head()
y_comp= df['Comparison']
y_comp.head()
y_prod= df['Prod Data']
y_prod.head()
y_cons= df['Cons Data']
y_cons.tail()
```

```python
from sklearn.preprocessing import scale
X_scaled = scale(X)
y_comp_scaled = scale(y_comp)
y_prod_scaled = scale(y_prod)
y_cons_scaled = scale(y_cons)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_scaled = sc.fit_transform(X)
y_comp_scaled = sc.fit_transform(y_comp.values.reshape(-1, 1))
y_prod_scaled = sc.fit_transform(y_prod.values.reshape(-1, 1))
y_cons_scaled = sc.fit_transform(y_cons.values.reshape(-1, 1))
```

```python
# Save predictors and its number to n_cols
n_cols = X_scaled.shape[1] 

# Set up the model
model = Sequential()

# Add the first layer
model.add(Dense(15, activation='relu', input_shape=(n_cols,), kernel_initializer='glorot_uniform'))

# Add the second layer
#model.add(Dense(20, activation='relu'))

# Add the output layer
model.add(Dense(1))

# Compile the model
adam = optimizers.Adam(lr=0.01)
model.compile(optimizer='adam', loss='mse', metrics=[R2])

# Fit the model
history = model.fit(X_scaled, y_comp_scaled, epochs=250, verbose=1, validation_split=0.3)

# Plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.set_title('Mean squared error')
ax1.set_xlabel('Epoch')

ax2.plot(history.history['R2'])
ax2.plot(history.history['val_R2'])
ax2.set_title('$R^2$')
ax2.set_xlabel('Epoch')

plt.tight_layout()
```
